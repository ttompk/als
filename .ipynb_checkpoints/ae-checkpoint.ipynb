## Adverse Events

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

##

```python
ae = pd.read_csv("data/AdverseEvents.csv")
```

```python
ae.head()
```

```python
# limit to only max dates
ae_max_date = ae.groupby('subject_id').max()
ae_max_date.reset_index(inplace=True)
ae_max_date['max_ae_start_date_study_day'] = ae_max_date['Start_Date_Delta']
ae_max_date['max_ae_end_date_study_day'] = ae_max_date['End_Date_Delta']
ae_max_date=ae_max_date[['subject_id', 'max_ae_start_date_study_day', 'max_ae_end_date_study_day']]
ae_max_date.head()
```

### Add Onset data

```python
# add the onset data
onset = pd.read_csv("clean_data/onset_date.csv")
```

```python
onset.drop(['delta_onset_diag', 'Diagnosis_Delta'], axis=1, inplace=True)
```

```python
onset.head()
```

```python
ae_max = ae_max_date.merge(onset, how='left', on='subject_id')
```

```python
ae_max.head()
```

```python
ae_max['max_ae_onset_start'] = np.abs(ae_max['Onset_Delta']) + ae_max['max_ae_start_date_study_day']
ae_max['max_ae_onset_end'] = np.abs(ae_max['Onset_Delta']) + ae_max['max_ae_end_date_study_day']
ae_max.drop(['max_ae_start_date_study_day','max_ae_end_date_study_day'], axis=1, inplace=True)
ae_max['max_ae_date_onset'] = ae_max[['max_ae_onset_start','max_ae_onset_end']].max(axis=1)
ae_max.drop(['max_ae_onset_start','max_ae_onset_end', 'Onset_Delta'], axis=1, inplace=True)
```

```python
ae_max.head()
```

```python
# save to csv
filename = 'clean_data/max_ae_date.csv'
ae_max_date.to_csv(filename, index=False)
```
